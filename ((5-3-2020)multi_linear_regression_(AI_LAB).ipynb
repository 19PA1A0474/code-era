```
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
from sklearn import metrics
```

```
j=pd.read_csv('students_placement_data.csv')
j
```

```
k=j.loc[j['Section']=='A']
k
```

```
j.isna().any()
```

```
j=j.fillna(method="ffill")
j
```

```
j.isna().any()
```

```
x = j[['SSC Percentage','inter_Diploma_percentage','Backlogs']].values
y= j['B.Tech_percentage'].values
```

```
x.shape
```

```
y.shape
```

```
X_train,X_test,Y_train,Y_test=train_test_split(x,y, test_size = 0.2 , random_state = 0)
```

```
model=LinearRegression()
model.fit(X_train,Y_train)
```

```
model.intercept_
```

```
model.coef_
```

```
predicted_btech=model.predict(X_test)
predicted_btech.shape
```

```
X_test.shape
```

```
print('R2 error:',metrics.r2_score(Y_test, predicted_btech))
```

```
print('R2 error:',metrics.mean_absolute_error(Y_test, predicted_btech))
```
